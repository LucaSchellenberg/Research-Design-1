```python
%cd /content/hifi-gan

import torch, json
from models import Generator
from utils import load_checkpoint
from env import AttrDict

# âœ… Correct path
with open("checkpoints/pretrained/LJ_V2/config.json") as f:
    h = AttrDict(json.load(f))

generator = Generator(h).to('cuda')
state_dict_g = torch.load("checkpoints/pretrained/LJ_V2/generator_v2", map_location='cuda')
generator.load_state_dict(state_dict_g['generator'])
generator.eval()
generator.remove_weight_norm()
```

```python
import numpy as np
import soundfile as sf

# Load mel spectrogram
mel = np.load("your_mel.npy")  # shape: [80, T]
mel_tensor = torch.FloatTensor(mel).unsqueeze(0).to('cuda')  # [1, 80, T]

# Generate audio
with torch.no_grad():
    audio = generator(mel_tensor).squeeze().cpu().numpy()

# Save output
sf.write("output.wav", audio, 22050)
```
